# Métodos especiales
* \_\_init\_\_\
  El método __init__ se usa para inicializar objetos de una clase. Se llama automáticamente cuando se crea una instancia de la clase y este define los atributos iniciales del objeto a partir de parámetros dados.
* \_\_str\_\_\
  El método \_\_str\_\_\ retorna un string que representa al objeto de un modo informal y facilmente interpretable por un usuario. Se llama automáticamente cuando se usa la función str() en un objeto (o sea: con str(objeto) y objeto.\_\_str\_\_\() se obtiene la misma cadena).
  Un ejemplo de este método está en la clase __Vehiculo__, donde se devuelven las caraterísticas del vehículo.

* \_\_repr\_\_\
  El método \_\_repr\_\_\, en similitud a \_\_str\_\_\, retorna un string que representa al objeto. Sin embargo, esta representación es detallada y no ambigua. No se apunta a que la lea un usuario, como \_\_str\_\_\, sino para desarrollo de programas.
* \_\_add\_\_\
  El método \_\_add\_\_ define la operación de suma. Se llama automáticamente cuando se usa el operador "+" entre dos objetos.
  Un ejemplo de este método está en la clase __Punto__ en que se define la suma entre puntos como la suma de componente a componente.
* \_\_sub\_\_\
  El método \_\_sub\_\_ define la operación de resta. Se llama automáticamente cuando se usa el operador "-" entre dos objetos.
* \_\_mul\_\_\
  El método \_\_mul\_\_ define la operación de multiplicación. Se llama automáticamente cuando se usa el operador "*" entre dos objetos.
* \_\_div\_\_\
  El método \_\_div\_\_ define la operación de división. Se llama automáticamente cuando se usa el operador "/" entre dos objetos.

Las figuras especificas, circulo, rectangulo y triangulo, son figuras y tienen reglas en su construcción que es necesario implementar. Usando la herencia, podes simplificar el comportamiento en perimetro y area pero la forma que tiene un `Cuadrado` (por ejemplo) es muy especifica.


```python
class Punto:
  '''
  La clase Punto crea la representación de un punto.
  
  Atributos:
      x (número entero o float): componente x del punto.
      y (número entero o float): componente y del punto.
  '''

  def __init__(self,componente_x,componente_y):
    '''
    El método __init__ define los atributos x e y asociados a las componentes de el punto.

    Argumentos:
        componente_x (número entero o float): componente x del punto.
        componente_y (número entero o float): componente y del punto.
    '''

    self.x = componente_x
    self.y = componente_y

  def distancia(self,otro_punto):
    '''
    El método distancia calcula la distancia entre dos puntos a partir de sus componentes.

    La distancia utilizada es la euclidea.
    
    Argumentos:
        otro_punto (instancia de tipo Punto): el otro punto con el cual se quiere calcular la distancia.

    Retorno:
        distancia (número entero o float): la distancia entre un punto y el otro.
    '''

    distancia = ((self.x-otro_punto.x)**2+(self.y-otro_punto.y)**2)**(1/2)
    return distancia

  def __add__(self,otro_punto):
    '''
    El método __add__ suma dos puntos.
    
    Argumentos:
        otro_punto (instancia de tipo Punto): punto que se suma.
        
    Retorno:
        suma (instancia de tipo Punto): la suma de los dos puntos.
    '''

    suma = Punto(self.x + otro_punto.x, self.y + otro_punto.y)
    return suma

  def convertir_a_punto(self,tupla_a_convertir):
    '''
    El método convertir_a_punto convierte una tupla en un punto.

    Argumentos:
        tupla_a_convertir (tuple): tupla con dos elementos representando las coordenadas x e y de un punto.

    Retorno:
        nuevo_punto (instancia de tipo Punto): el punto correspondiente a la tupla.
    '''

    nuevo_punto = Punto(tupla_a_convertir[0],tupla_a_convertir[1])
    return nuevo_punto 

class Figura:
  '''
  La clase Figura crea la representación de una figura a partir de las coordenadas de sus vértices.

  Atributos:
      vertices (lista de instancias de tipo Punto): la lista de puntos que identifican a los vértices de la figura.
  '''

  def __init__(self,vertices):
    '''
    El método __init__ inicializa el atributo vertices. 

    Los vertices deben estar en orden, horario o antihorario, de lo contrario los métodos de la clase no funcionarán correctamente.

    Argumentos:
        vertices (lista de instancias de tipo Punto): la lista de puntos que identifican a los vértices de la figura.
    '''

    self.vertices = vertices

  def calcular_area(self):
    '''
    El método calcular_area calcula el área de la figura con la técnica de Gauss.
    '''

    i=0
    termino_pos = 0
    termino_neg = 0
    while i + 1 < len(self.vertices):
        termino_pos = termino_pos + (self.vertices[i].x)*(self.vertices[i+1].y)
        termino_neg = termino_neg - (self.vertices[i+1].x)*(self.vertices[i].y)
        i = i + 1
    
    sumatorio = (termino_pos + termino_neg)
    area = (1/2)*abs(sumatorio) 
    return area

  def calcular_perimetro(self):
    '''
    calcular_perimetro calcula el perímetro de la figura.
    
    Retorno:
        perimetro (número entero o float): el perímetro de la figura.
    '''

    cdd_vertices = len(self.vertices)
    vertices_ampliados = self.vertices.append(self.vertices[0])
    i = 0
    perimetro = 0
    while i < cdd_vertices:
        perimetro = perimetro + self.vertices[i].distancia(self.vertices[i+1])
        i = i + 1
    
    return perimetro

class Rectangulo(Figura):
  '''
  La clase rectángulo crea la representación de un rectángulo.

  Atributos:
      vertices (lista de instancias de tipo Punto): la lista de puntos que identifican a los vértices del rectángulo.
  '''

  def __init__(self,vertices):
    '''
    El método __init__ inicializa la instancia.
  
    Argumentos:
        vertices (lista de instancias de tipo Punto): la lista de puntos que identifican a los vértices del rectángulo.
        Los vertices deben estar en orden, horario o antihorario, de lo contrario los métodos de la clase no funcionarán correctamente.
        
    '''

    super().__init__(vertices)
    
  def calcular_area(self):
    '''
    El método calcular_area calcula el área del rectángulo.

    Retorno:
        area (número): el área del rectángulo.
    '''

    area = (self.vertices[0].distancia(self.vertices[1]))*(self.vertices[1].distancia(self.vertices[2]))
    return area   

  def calcular_perimetro(self):
    '''
    El método calcular_perimetro calcula el perímetro del rectángulo.

    Retorno:
        perimetro (número): el perímetro del rectángulo.
    '''

    perimetro = 2*((self.vertices[0].distancia(self.vertices[1]))+(self.vertices[1].distancia(self.vertices[2])))
    return perimetro
     
class Cuadrado(Figura):
  '''
  La clase cuadrado crea una representación de un cuadrado.

  Atributos:
      vertices (lista de instancias de tipo Punto): la lista de puntos que identifican a los vértices del cuadrado.
  '''

  def __init__(self,vertices):
    '''
    El método __init__ inicializa la instancia.

    Argumentos:
        vertices (lista de instancias de tipo Punto): la lista de puntos que identifican a los vértices del cuadrado.
        Los vertices deben estar en orden, horario o antihorario, de lo contrario los métodos de la clase no funcionarán correctamente.
    '''
    super().__init__(vertices)

  def calcular_area(self):
    '''
    El método calcular_area calcula el área del cuadrado.

    Retorno:
        area (número): el área del cuadrado.
    '''

    area = (self.vertices[0].distancia(self.vertices[1]))**2
    return area

  def calcular_perimetro(self):
    '''
    El método calcular_perimetro calcula el perímetro del cuadrado.

    Retorno:
        perimetro (número): el perímetro del cuadrado.
    '''

    perimetro = 4*(self.vertices[0].distancia(self.vertices[1]))
    return perimetro

class Circulo(Figura):
  '''
  La clase círculo crea una representación de un círculo.

    Atributos:
      vertices (lista de instancias de tipo Punto): la lista de puntos que identifican al círculo.
      El primer punto corresponde al centro del cículo, el segundo puede ser cualquiera que pertenezca a la circunferencia.
  '''


  def __init__(self):
    '''
    El método __init__ inicializa la instancia.

    Argumentos:
        vertices (lista de instancias de tipo Punto): la lista de puntos que identifican al círculo.
        El primer punto corresponde al centro del cículo, el segundo puede ser cualquiera que pertenezca a la circunferencia.        
    '''

    super().__init__(self)
    #self.radio = self.vertices[0].distancia(self.vertices[1]) ¿Corresponde o no corresponde?
                                                              #Porque es la característica principal del círculo, y se usa en sus método.
                                                              #Si no, lo dejo como método "radio".

  def calcular_area(self):
    '''
    El método calcular_area calcula el área del círculo.

    Para el cálculo del área se utiliza pi con 5 cifras significativas.

    Retorno:
        area (número): el área del círculo.
    '''

    pi = 3.1415
    area = pi*(self.vertices[0].distancia(self.vertices[1]))**2
    return area

  def calcular_perimetro(self):
    '''
    El método calcular_perimetro calcula el perímetro del círculo.

    Para el cálculo del perímetro se utiliza pi con 5 cifras significativas.

    Retorno:
        perimetro (número): el perímetro del círculo.
    '''

    pi = 3.1415
    perimetro = 2*pi*(self.vertices[0].distancia(self.vertices[1]))
    return perimetro

  def le_pertenece(self, otro_punto):
    '''
    El método le_pertenece verifica si un punto le pertenece a la circunferencia del círculo.

    Argumentos:
        otro_punto (instancia Punto): el punto que se quiere saber si pertenece a la circunferencia. 

    Retorno:
        pertenece (booleano): True si el punto pertece a la circunferencia.
    '''
    
    pertenece = self.vertices[0].distancia(self.vertices[1]) == self.vertices[0].distancia(otro_punto)
    return pertenece

if __name__ == "__main__":

  mis_vertices = [Punto(0,0),Punto(5,0),Punto(5,5),Punto(2.5,6),Punto(0,5)]  
  mi_figura = Figura(mis_vertices)
  mi_area = mi_figura.calcular_area()
  print(f"{mi_area}")
  mi_perimetro = mi_figura.calcular_perimetro() 
  print(mi_perimetro) 
  mi_cuadrado = Cuadrado([Punto(0,0),Punto(5,0),Punto(5,5),Punto(0,5)])
  area_cuadrado = mi_cuadrado.calcular_area()
  print(area_cuadrado)
  mi_circulo = Circulo([Punto(0,1),Punto(0,1)])
  otro_punto = Punto(0,5)
  pertenencia = mi_circulo.le_pertenece(otro_punto)
  print(pertenencia)
```

el método `cambiar` introduce inconsistencias en la representación interna de la `Hora`, es posible cambiar a segundos y minutos mayor a 60. Y el atributo `segundos_totales` no es modificado junto a los otros.

```python
# Tiempo
# Crear la clase Hora que represente un momento, definido como hora, minuto y segundo. Agregar métodos para facilitar la suma entre estas clases, comparaciones y representación como cadena.

class Hora: #Solo usar segundos, simplifica métodos de la clase.
  '''
  La clase Hora crea la representación de una hora a partir de la cantidad de segundos totales.

  El sistema horario utilizado es de 24 horas. 

  Atributos
  ---------
    segundos_totales : número entero
        la hora expresada únicamente en segundos.
    '''

  def __init__(segundos_totales=0):
    '''
    __init__ inicializa la instancia. 

    Si no se da el argumento, este toma el valor por defecto 0.
    En caso de que los valores de los argumentos se excedan del intervalo definido en el sistema horario para los atributos homónimos, estos son ajustados.

    Parámetros
    ----------
      segundos_totales : número entero
         la hora expresada únicamente en segundos.
    '''    

    self.segundos_totales = segundos_totales

  def __add__(self,otra_hora):
    '''
    El método __add__ suma dos horas. 

    Argumentos
    ----------
      otra_hora : instancia de tipo Hora
          la otra hora que se quiere sumar

    Retorno
    -------
      suma : instancia de tipo Hora
          la suma resultante
    '''
    suma_segundos_totales = self.segundos_totales + otra_hora.segundos_totales
    suma = Hora(segundos=suma_segundos_totales)
    return suma

  def es_igual_a(self,otra_hora):
    '''
    es_igual compara si dos horas son iguales.
    
    Argumentos
    ----------
      otra_hora : instancia de tipo Hora
          la otra hora que se quiere sumar

    Retorno
    -------
      condicion_total : booleano
          True si las dos horas son iguales, False si son distintas
    '''
    condicion_hora = self.horas == otra_hora.horas
    condicion_minuto = self.minutos == otra_hora.minutos
    condicion_segundo = self.segundos == otra_hora.segundos
    condicion_total = condicion_hora and condicion_minuto and condicion_segundo
    return condicion_total

  def cambiar(self,nuevas_horas,nuevos_minutos,nuevos_segundos):
    '''
    cambiar modifica los atributos horas, minutos y segundos.
    
    Argumentos
    ----------
      nuevas_horas : número entero
          el nuevo valor de horas 

      nuevos_minutos : número entero
          el nuevo valor de minutos

      nuevos_segundos : número entero
          el nuevo valor de segundos                 
    '''
    self.horas = nuevas_horas
    self.minutos = nuevos_minutos
    self.segundos = nuevos_segundos
    # Operar análogamente a como hice en __init__ para ajustar las horas, minutos y segundos. Así como está es posible cambiar a minutos y horas > 60!!!
    # Modificar segundos_totales en conjunto y coherencia con el resto de los atributos.
    
  def __str__(self):
    '''
    __str__ genera una cadena que representa la hora.
    
    Retorno
    -------
      string con los valores de los atributos de la hora, indicados con abreviaturas
    '''
    return f"{self.horas}hs{self.minutos}min{self.segundos}seg"

if __name__ == "__main__":
  mi_hora = Hora(24,20,30)
  otra_hora = Hora(2,30,40)
  total = mi_hora + otra_hora
  string_total = str(total)
  print(f"{string_total=}")
  string_mi_hora = str(mi_hora)
  print(f"{string_mi_hora=}")
  mi_hora.cambiar(9,8,7)
  print(mi_hora)
  comparacion = mi_hora.es_igual_a(otra_hora)
  if comparacion:
    print(f"{mi_hora} es igual a {otra_hora}")
  else:
    print(f"{mi_hora} no es igual a {otra_hora}")
```

#Vehículos 
_Creen las clases necesarias para representar Vehiculos desde el punto de vista de un usuario, incluyendo los métodos y atributos necesarios Crear las clases derivadas Auto, Moto y Bicicleta._

```python
# Vehículos
# Creen las clases necesarias para representar Vehiculos desde el punto de vista de un usuario, incluyendo los métodos y atributos necesarios. Crear las clases derivadas Auto, Moto y Bicicleta.
class Vehiculo:
    '''
    La clase Vehiculo crea la representación de un vehículo.

     Atributos:
        marca (string): la marca del vehículo.
        color (string): el color del vehículo.
        cantidad_nafta (número entero o float): la cantidad de nafta.
    '''
    def __init__(self,marca="Desconocida",color="Gris (base)",cantidad_nafta=0):
        '''
        __init__ inicializa los atributos de la instancia: marca, color, cantidad_nafta.
        
        En caso de no ingresarse uno o varios argumentos, los atributos que dependen de éstos toman valores por defecto. 
        
        Argumentos:
          marca (string): la marca del vehículo.
          color (string): el color del vehículo.
          cantidad_nafta (número entero o float): la cantidad de nafta.
        '''
        self.marca = marca
        self.color = color
        self.cantidad_nafta = cantidad_nafta    
    
    def pintar(self,nuevo_color): 
        '''
        pintar redefine el atributo color con el nuevo valor nuevo_color.

        Argumentos:
          nuevo_color (string): el nuevo color del vehículo.
        '''
        self.color = nuevo_color
        
    def cargar(self,cantidad_a_cargar):
        '''
        cargar redefine el atributo cantidad_nafta de la instancia aumentandola en una cantidad igual a cantidad_a_cargar.
        
        Si cantidad_a_cargar es negativo, se iguala a 0. 

        Argumentos:
          cantidad_a_cargar (número  entero o float): la cantidad de nafta a cargar.
        '''
        if cantidad_a_cargar < 0:
          cantidad_a_cargar = 0
        self.cantidad_nafta = self.cantidad_nafta + cantidad_a_cargar

    def __str__(self):
      '''
      __str__ genera y arroja un string que contiene los atributos de la instancia.

      Retorno (string): un string donde se indica cada atributo de la instancia y su valor.
      '''
      string_vehiculo = f"Marca:{self.marca};Color:{self.color};Cantidad de nafta:{self.cantidad_nafta}"
      return string_vehiculo

class Auto(Vehiculo):
  '''
  La clase Auto crea la representación de un auto.
  
  Auto es una subclase clase Vehiculo.

  Atributos:
        marca (string): la marca del vehículo.
        color (string): el color del vehículo.
        cantidad_nafta (número entero o float): la cantidad de nafta.
        puertas_total (número entero): cantidad de puertas del auto.
        puertas_abiertas (número entero): cantidad de puertas abiertas.
  '''
  def __init__(self,puertas_total=4,puertas_abiertas=0,marca="Desconocida",color="Gris (base)",cantidad_nafta=0):
    '''
    El método __init__ inicializa los atributos de la instancia propias del tipo Auto (puertas_total,puertas_abiertas) y las de la clase-padre Vehiculo (marca, color, cantidad_nafta).
    
    En caso se que no se ingresen un argumentos, estos toman valores por defecto.

    Argumentos:
        marca (string): la marca del vehículo.
        color (string): el color del vehículo.
        cantidad_nafta (número entero o float): la cantidad de nafta.
        puertas_total (número entero): cantidad de puertas del auto.
        puertas_abiertas (número entero): cantidad de puertas abiertas.
    '''
    super().__init__(marca,color,cantidad_nafta) 
    self.puertas_total = puertas_total
    self.puertas_abiertas = puertas_abiertas

  def __str__(self):
    string_auto = super().__str__() + f";Total de puertas:{self.puertas_total};Puertas abiertas:{self.puertas_abiertas}"
    return string_auto

class Moto(Vehiculo):
  '''
  La clase Moto crea la representación de una Moto.
  
  Moto es una subclase clase Vehiculo.

   Atributos:
        marca (string): la marca del vehículo.
        color (string): el color del vehículo.
        cantidad_nafta (número entero o float): la cantidad de nafta.
        estado_cadena (string): el estado de la cadena.  
  '''
  def __init__(self,estado_cadena="Vieja",marca="Desconocida",color="Gris (base)",cantidad_nafta=0):
    '''
    __init__ inicializa los atributos de la instancia propias del tipo Moto (estado_cadena) y las de la clase-padre Vehiculo (marca, color, cantidad_nafta).
    
    En caso se que no se ingresen un argumentos, estos toman valores por defecto.

   Argumentos:
        marca (string): la marca del vehículo.
        color (string): el color del vehículo.
        cantidad_nafta (número entero o float): la cantidad de nafta.
        estado_cadena (string): el estado de la cadena.
    '''
    super().__init__(marca,color,cantidad_nafta) 
    self.estado_cadena = estado_cadena
  
  def cambiar_cadena(self):
    '''
    cambiar_cadena cambia la cadena de la moto.

    Redefine el atributo estado_cadena a "Nueva".
    '''
    self.estado_cadena = "Nueva"

  def __str__(self):
    string_moto = super().__str__() + f";Estado de la cadena:{self.estado_cadena}"
    return string_moto    

class Bici(Vehiculo):
  '''
  La clase Bici crea la representación de una Bici.

  Bici es una subclase clase Vehiculo.

   Atributos:
        marca (string): la marca del vehículo.
        color (string): el color del vehículo.
        cantidad_nafta (número entero o float): la cantidad de nafta.
        movimiento (string): el estado de movimiento.
  '''
  def __init__(self,movimiento="Detenida",marca="Desconocida",color="Gris (base)",cantidad_nafta=0):
    '''
    __init__ inicializa los atributos de la instancia propias del tipo Bici (movimiento) y las de la clase-padre Vehiculo (marca, color, cantidad_nafta).
    
     En caso se que no se ingresen un argumentos, estos toman valores por defecto.

   Argumentos:
        marca (string): la marca del vehículo.
        color (string): el color del vehículo.
        cantidad_nafta (número entero o float): la cantidad de nafta.
        movimiento (string): el estado de movimiento.
    '''
    super().__init__(marca,color,cantidad_nafta=0)
    self.movimiento = movimiento

  def pedalear(self):
    '''
    pedalear pone en movimiento a la bici.

    Redefine el atributo movimiento a "En movimiento".
    '''
    self.movimiento = "En movimiento"

  def frenar(self):
    '''
    frenar frena a la bici.

    Redefine el atributo movimiento a "Detenida".    
    '''
    self.movimiento = "Detenida"

  def __str__(self):
    string_bici = super().__str__() + f";Estado de movimiento:{self.movimiento}"
    return string_bici

if __name__ == "__main__":
  auto_Anna = Auto()
  print(auto_Anna.color)
  auto_Anna.pintar("azul")
  print(str(auto_Anna))

  moto_Juan = Moto("Vieja",4)
  print(str(moto_Juan))
  moto_Juan.cambiar_cadena()
  print(str(moto_Juan))

  bici_Fran = Bici("Detenida")
  print(str(bici_Fran))
  bici_Fran.pedalear()
  print(str(bici_Fran))
```
